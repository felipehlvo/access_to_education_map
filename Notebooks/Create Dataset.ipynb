```python
import pandas as pd
import os
import numpy as np
from glob import glob
```

```python

# assign directory
from ast import Raise
import csv


directory = '/Users/feliphlvo/Documents/Minerva/Capstone/data/Censo 2010'
n = 1

# Get a list of the table names
table_names = [i.split("_")[0] for i in os.listdir("/Users/feliphlvo/Documents/Minerva/Capstone/data/Censo 2010/AC/Base informaçoes setores2010 universo AC/CSV")]
print(table_names)

# For each table name
for table in table_names:
    df = pd.DataFrame()
    # iterate over folders in the data
    for state in os.listdir(directory):
        f = os.path.join(directory, state)
        # checking if it is a folder
        if os.path.isdir(f):
            print(n)
            print(state)
            n += 1
            # For each folder (representing each state)
            for folder in os.listdir(f):
                # Open the csv file of the basic information about the census tracts of that state
                f2 = os.path.join(f, folder)
                if os.path.isdir(f2):

                    try:
                        csv_file = glob(f2 + f"/CSV/?{table[1:]}*.csv")[0]
                    except:
                        
                        raise Exception(state, table)
                        
                    print(csv_file)

                    state = pd.read_csv(csv_file, sep=";")
                    df = df.append(state)
                    print(df.shape)
    df.to_csv(f"/Users/feliphlvo/Documents/Minerva/Capstone/data/{table}_combined.csv")
```
